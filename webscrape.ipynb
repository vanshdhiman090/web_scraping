```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
url = "https://www.flipkart.com/search?q=i+phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/115.0.0.0 Safari/537.36"
}



response = requests.get(url, headers=header)
soup = BeautifulSoup(response.content, "html.parser")
display(response)
```

```python

name = []
price = []
discount = []
rom = []
size = []
processor = []
camera = []
rating = []

for page in range(1,25):


 soup = BeautifulSoup(response.text, 'lxml')
 Container = soup.find_all('div', class_="tUxRFH")

 for item in Container :
   iphone_name = item.find('div', class_="KzDlHZ")
   iphone_price = item.find('div', class_="Nx9bqj _4b5DiR")
   iphone_rating = item.find('div', class_="XQDdHH")
   iphone_discount = item.find('div', class_="UkUFwK")
   iphone_features = item.find_all('li' , class_ ="J+igdf")
   iphone_ram = iphone_features[0]
   iphone_size = iphone_features[1]
   iphone_camera = iphone_features[2]
   iphone_processor = iphone_features[3]

   Iphone_name = iphone_name.text.strip() if iphone_name else None
   Iphone_rating = iphone_rating.text.strip() if iphone_rating else None
   Iphone_discount = iphone_discount.text.strip() if iphone_discount else None
   Iphone_price = iphone_price.text.strip() if iphone_price else None
   Iphone_ram = iphone_ram.text.strip() if iphone_ram else None
   Iphone_size = iphone_size.text.strip() if iphone_size else None
   Iphone_camera = iphone_camera.text.strip() if iphone_camera else None
   Iphone_processor = iphone_processor.text.strip() if  iphone_processor else None

   name.append(Iphone_name)
   price.append(Iphone_price)
   discount.append(Iphone_discount)
   rom.append(Iphone_ram)
   size.append(Iphone_size)
   processor.append(Iphone_processor)
   camera.append(Iphone_camera)
   rating.append(Iphone_rating)
 time.sleep(3)
data = pd.DataFrame({
    'Iphone Name' : name,
    'Price' : price,
    'Rating' : rating,
    'Discount' : discount,
    'Rom' : rom,
    'Size' : size,
    'Camera' : camera,
    'Processor' : processor
})
display(data)

```
